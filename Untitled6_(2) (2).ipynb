```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
```

```python
# Load the dataset
file_path = "Copy of Student Academic form (Responses).xlsx"
df = pd.read_excel(file_path)

print("Columns in dataset:")
print(df.columns)
df.head()
```

```python
# Text GPA column
gpa_col = "What is your current GPA"  # make sure this matches exactly

# Map GPA bands to numeric midpoints
gpa_map = {
    "0-40": 20,
    "40-50": 45,
    "50-60": 55,
    "60-70": 65,
    "70+": 75
}
df["GPA_num"] = df[gpa_col].map(gpa_map)

# Binary target: High_GPA = 1 if 60-70 or 70+, else 0
high_gpa_bands = ["60-70", "70+"]
df["High_GPA"] = df[gpa_col].apply(lambda x: 1 if x in high_gpa_bands else 0)

df[[gpa_col, "GPA_num", "High_GPA"]].head()
```

```python
df = df.drop(columns=[
    "Timestamp",
    "@dropdown",
    "Within college do you feel there is someone you can talk to about your day-to-day problems"
], errors="ignore")

df.head()
```

```python
# Target
y = df["High_GPA"]

# All other columns except target + original GPA text
X_raw = df.drop(columns=["High_GPA", gpa_col, "GPA_num"])


# One-hot encode all categorical columns
X_encoded = pd.get_dummies(X_raw, drop_first=True)

# Save feature names (used later for interactive predictor)
feature_cols = X_encoded.columns

X_encoded.head()
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y,
    test_size=0.2,
    random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
# Train logistic regression model
model = LogisticRegression(max_iter=2000)
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Accuracy + classification report
acc = accuracy_score(y_test, y_pred)
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)

print("Test accuracy:", acc)
print("\nClassification report:")
print(classification_report(y_test, y_pred))
```

```python
def predict_student_gpa_interactive():
    print("Answer these questions about yourself:\n")

    # We build a dict that looks like ONE ROW of your original df (X_raw)
    new_row = {}

    # 1) Year of study
    print("What year are you currently in?")
    print("1: 1st Year")
    print("2: 2nd Year")
    print("3: 3rd Year")
    print("4: 4th Year")
    year_choice = int(input("Choose 1-4: "))
    year_map = {
        1: "1st Year",
        2: "2nd Year",
        3: "3rd Year",
        4: "4th Year"
    }
    new_row["What year are you currently in?"] = year_map.get(year_choice, "1st Year")

    # 2) Study hours
    print("\nOn average how many hours a week do you spend studying?")
    print("1: 0-2")
    print("2: 2-4")
    print("3: 4-6")
    print("4: 6+")
    study_choice = int(input("Choose 1-4: "))
    study_map = {
        1: "0-2",
        2: "2-4",
        3: "4-6",
        4: "6+"
    }
    new_row["On average how many hours a week do you spend studying"] = study_map.get(study_choice, "2-4")

    # 3) Student accommodation
    accom_ans = input("\nDo you have student accommodation? (Yes/No): ").strip().title()
    if accom_ans not in ["Yes", "No"]:
        accom_ans = "No"
    new_row["Do you have student accommodation?"] = accom_ans

    # Walking distance (only really matters if Yes, but we still pick a category)
    print("\nIf you have accommodation, how far walking is your accommodation from the college?")
    print("1: 0-10 minutes")
    print("2: 10-30 minutes")
    print("3: 30 minutes-1 hour")
    print("4: 1 hour+")
    walk_choice = int(input("Choose 1-4 (if No accom, just pick closest): "))
    walk_map = {
        1: "0-10 minutes",
        2: "10-30 minutes",
        3: "30 minutes-1 hour",
        4: "1 hour+"
    }
    new_row["If so, how far walking is your accommodation from the college?"] = walk_map.get(walk_choice, "10-30 minutes")

    # 4) Commute
    commute_ans = input("\nDo you commute to college (outside of accom)? (Yes/No): ").strip().title()
    if commute_ans not in ["Yes", "No"]:
        commute_ans = "No"
    new_row["Do you commute to college"] = commute_ans

    print("\nIf you commute, how long is your commute?")
    print("1: No Commute")
    print("2: 0-30 minutes")
    print("3: 30 minutes- 1 hour")
    print("4: 1-2 hours")
    print("5: 2+ hours")
    commute_choice = int(input("Choose 1-5: "))
    commute_map = {
        1: "No Commute",
        2: "0-30 minutes",
        3: "30 minutes- 1 hour",
        4: "1-2 hours",
        5: "2+ hours"
    }
    new_row["If so, how long is your commute?"] = commute_map.get(commute_choice, "0-30 minutes")

    # 5) Extracurriculars
    extra_ans = input("\nAre you involved in extracurricular activities? (Yes/No): ").strip().title()
    if extra_ans not in ["Yes", "No"]:
        extra_ans = "No"
    new_row["Are you involved in extracurricular activities"] = extra_ans

    print("\nHow many hours do you spend doing extracurricular activities a week?")
    print("1: No Extracurrcular")
    print("2: 0-2")
    print("3: 2-4")
    print("4: 4-6")
    print("5: 6+")
    extra_choice = int(input("Choose 1-5: "))
    extra_map = {
        1: "No Extracurrcular",
        2: "0-2",
        3: "2-4",
        4: "4-6",
        5: "6+"
    }
    new_row["How many hours do you spend doing extracurricular activities a week"] = extra_map.get(extra_choice, "0-2")

    # 6) Employment
    employed_ans = input("\nAre you currently employed? (Yes/No): ").strip().title()
    if employed_ans not in ["Yes", "No"]:
        employed_ans = "No"
    new_row["Are you currently employed?"] = employed_ans

    print("\nIf employed, how many hours a week do you work?")
    print("1: 0 hours")
    print("2: 0-8 hours")
    print("3: 8-12 hours")
    print("4: 12-20 hours")
    print("5: 20+ hours")
    work_choice = int(input("Choose 1-5: "))
    work_map = {
        1: "0 hours",
        2: "0-8 hours",
        3: "8-12 hours",
        4: "12-20 hours",
        5: "20+ hours"
    }
    new_row["If so, how many hours a week do you work?"] = work_map.get(work_choice, "0-8 hours")

    # 7) Sleep
    print("\nHow many hours on average do you sleep per school night?")
    print("1: 2-4")
    print("2: 4-6")
    print("3: 6-8")
    print("4: 8+")
    sleep_choice = int(input("Choose 1-4: "))
    sleep_map = {
        1: "2-4",
        2: "4-6",
        3: "6-8",
        4: "8+"
    }
    new_row["How many hours on average do you sleep per school night"] = sleep_map.get(sleep_choice, "6-8")

    # 8) Screen time
    print("\nHow many hours of screen time do you average a day?")
    print("1: 0-2")
    print("2: 2-4")
    print("3: 4-6")
    print("4: 6-8")
    print("5: 8+")
    screen_choice = int(input("Choose 1-5: "))
    screen_map = {
        1: "0-2",
        2: "2-4",
        3: "4-6",
        4: "6-8",
        5: "8+"
    }
    new_row["How many hours of screen time do you average a day"] = screen_map.get(screen_choice, "4-6")

    # 9) Stress (numeric already)
    stress = int(input("\nOn a scale of 1-10 how stressed would you consider yourself? (1-10): "))
    new_row["On a scale of 1-10 how stressed would you consider yourself"] = stress

    # --- Convert this single row into same encoded format as training data ---
    X_new_raw = pd.DataFrame([new_row])
    X_new_encoded = pd.get_dummies(X_new_raw, drop_first=True)

    # Align columns with training feature columns (missing ones filled with 0)
    X_new_encoded = X_new_encoded.reindex(columns=feature_cols, fill_value=0)

    # Predict
    prob_high = model.predict_proba(X_new_encoded)[0, 1]
    pred_class = model.predict(X_new_encoded)[0]

    print("\n====================================")
    print(f"Predicted probability of GPA 60+ : {prob_high:.1%}")
    if pred_class == 1:
        print("Prediction: You are LIKELY to score 60 or above.")
    else:
        print("Prediction: You are UNLIKELY to score 60 or above.")
    print("====================================")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure df is defined with 'High_GPA' for plotting
# Re-load the dataset and preprocess df to ensure 'High_GPA' is available
file_path = "/content/Copy of Student Academic form (Responses).xlsx"
df = pd.read_excel(file_path)

gpa_col = "What is your current GPA"
high_gpa_bands = ["60-70", "70+"]
df["High_GPA"] = df[gpa_col].apply(lambda x: 1 if x in high_gpa_bands else 0)

# Percentage of High_GPA per commute time band
commute_col = "If so, how long is your commute?"

# Handle '0' values and NaN values before converting to string for plotting
df[commute_col] = df[commute_col].replace(0, 'No Commute') # Replace numeric 0
df[commute_col] = df[commute_col].fillna('No Commute') # Fill NaN values
df[commute_col] = df[commute_col].astype(str) # Ensure all are strings for consistent plotting

high_pct_by_commute = (df.groupby(commute_col)["High_GPA"]
                       .mean()
                       .sort_values(ascending=False) * 100)

plt.figure(figsize=(10, 6))
plt.bar(high_pct_by_commute.index, high_pct_by_commute.values)
plt.xticks(rotation=45, ha="right")
plt.ylabel("% of Students with GPA 60+")
plt.title("Percentage of High GPA (60+) by Commute Time")
plt.tight_layout()
plt.show()
```
